```python
# Primeiro vou separar a coluna "order_purchase_timestamp" em data e mês e passar as novas colunas criadas para datetime com o to_datetime

df_pedidos['purchase_date'] = df_pedidos['order_purchase_timestamp'].dt.date
df_pedidos['purchase_month'] = df_pedidos['order_purchase_timestamp'].dt.strftime('%Y-%m-01') #mantive o 01 em dd indicado pela professora na aula
df_pedidos['purchase_date'] = pd.to_datetime(df_pedidos['purchase_date'])
df_pedidos['purchase_month'] = pd.to_datetime(df_pedidos['purchase_month'])
df_pedidos
```

```python
#Meus gráficos puxando direto do df_pedidos não estavam funcionando, então decidi criar um novo DataFrame só dos pedidos cancelados

df_pedidos_cancelados = df_pedidos[df_pedidos.order_status == "canceled"] 
df_pedidos_cancelados
```

```python
#Verifiquei os tipos pra ver se permanecia como no df_pedidos

df_pedidos_cancelados.dtypes
```

```python
#Peguei a ideia que a professora deu em sala e repliquei, fazendo um recorte bem parecido, mas sobre os pedidos cancelados apenas

df_sazonalidade = df_pedidos_cancelados.groupby('purchase_month').agg({'order_id': 'nunique'})
df_sazonalidade.sort_values(by='order_id', ascending=False)
```

```python
#Com isso foi possível plotar o gráfico que eu estava imaginando

plt.figure(figsize=(20,6))
sns.lineplot(
    data=df_sazonalidade.reset_index().sort_values("purchase_month"),
    y="order_id",
    x="purchase_month",
    marker="o")

plt.title('Sazonalidade Pedidos Cancelados por Mês', fontsize=16)
plt.xlabel('Mês de Compra', fontsize=14)
plt.ylabel('Número de Pedidos Cancelados (order_id)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(True)

plt.show()


```

<p>3. Como é a distribuição das vendas em relação  às avaliações?</li>

<span style="color: green;">
Aqui, trabalhei em pair programming com uma colega, porque nossa ideia era avaliar se uma avaliação negativa ou positiva tem impacto sobre as vendas de um produto, mas a nossa ideia inicial não deu muito certo e tivemos que fazer adaptações.
Para isso foi necessário fazer uma junção entre os datasets: "olist_order_reviews_dataset" - que traz as avaliações + "olist_orders_dataset" - que traz informações das vendas + "olist_order_items" - que correlaciona o id do produto com o id do pedido.
<span>

```python
#Listando as colunas do df_pedidos para otimizar o drop
df_pedidos.columns
```

```python
#nosso df_pedidos já foi chamado acima e já tratamos a data da compra que é a principal informação que vamos usar, então, em um primeiro momento não vou precisar trabalhar nele, então posso dropar as colunas que não vou usar

columns_pedidos_drop = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date','order_estimated_delivery_date']
df_vendas_versus_avaliação  = df_pedidos.drop(columns_pedidos_drop, axis=1)
df_vendas_versus_avaliação
```

```python
df_avaliacoes_tratado = pd.read_csv("../../dataset_olist/dados brutos/olist_order_reviews_dataset.csv")
df_avaliacoes_tratado.columns
```

```python
#agora vamos começar a mergear com as informações úteis dos outros datasets, mas primeiro dropando deles o que não é últil para essa análicense

columns_avaliacoes_drop = ['review_comment_title', 'review_comment_message', 'review_answer_timestamp']
df_avaliacoes_tratado = df_avaliacoes_tratado.drop(columns_avaliacoes_drop, axis=1)
df_avaliacoes_tratado.dtypes
```

```python
#passando a data para datetime
df_avaliacoes_tratado['review_creation_date'] = pd.to_datetime(df_avaliacoes_tratado['review_creation_date'])
df_avaliacoes_tratado['review_date'] = df_avaliacoes_tratado['review_creation_date'].dt.date
df_avaliacoes_tratado['review_month'] = df_avaliacoes_tratado['review_creation_date'].dt.strftime('%Y-%m-%01')
df_avaliacoes_tratado['review_date'] = pd.to_datetime(df_avaliacoes_tratado['review_date'])
df_avaliacoes_tratado['review_month'] = pd.to_datetime(df_avaliacoes_tratado['review_month'])
df_avaliacoes_tratado.dtypes



```

```python
#agora vamos fazer o primeiro merge entre elas, pelo order_id

df_vendas_versus_avaliação = pd.merge(df_vendas_versus_avaliação, df_avaliacoes_tratado, on='order_id', how='inner')
df_vendas_versus_avaliação
```

```python
#agora vamos fazer o tratamento da tabela que permite relacionar produto com pedido

df_produtos_tratado = pd.read_csv('../../dataset_olist/dados brutos/olist_order_items_dataset.csv')
df_produtos_tratado.columns
```

```python
columns_produtos_drop = ['order_item_id', 'seller_id',
       'shipping_limit_date', 'price', 'freight_value']

df_produtos_tratado = df_produtos_tratado.drop(columns_produtos_drop, axis=1)
df_produtos_tratado.dtypes
```

```python
#Vamos mergear agora esse df para unificar e começar as análises

df_vendas_versus_avaliação = pd.merge(df_vendas_versus_avaliação, df_produtos_tratado, on='order_id', how='inner')
df_vendas_versus_avaliação
```

```python
#Vou criar uma coluna de classificação (review_classification) com base na nota da avaliação (review_score)

df_vendas_versus_avaliação.loc[df_vendas_versus_avaliação['review_score'] <=2, 'review_classification'] = 'Ruim'
df_vendas_versus_avaliação.loc[df_vendas_versus_avaliação['review_score'] == 3, 'review_classification'] = 'Neutro'
df_vendas_versus_avaliação.loc[df_vendas_versus_avaliação['review_score'] >=4, 'review_classification'] = 'Boa'
df_vendas_versus_avaliação
```

```python
df_vendas_versus_avaliação.columns
```

```python
df_vendas_versus_avaliação.info()
```

```python
df_vendas_versus_avaliação.to_csv("datasets_criados/df_vendas_versus_avaliacao.csv", index=False)
```

```python
#Aqui usei um pouco o chatGPT para ajudar a entender a lógica a ser aplicada para construir esse novo DF para fazer a avaliação
# 1. Agrupar os dados por 'review_classification' e contar as vendas (order_id)
df_resumo = df_vendas_versus_avaliação.groupby('review_classification')['order_id'].count().reset_index()

# 2. Renomear a coluna para 'total_vendas' para mais clareza
df_resumo.columns = ['review_classification', 'total_vendas']

# 3. Plotar o gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(df_resumo['review_classification'], df_resumo['total_vendas'], color=['green', 'orange', 'red'])
plt.xlabel('Classificação de Avaliação')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Classificação de Avaliação')
plt.xticks(rotation=0)
plt.show()
```

<div style="border: 2px solid #f0ad4e; padding: 10px; border-radius: 5px; background-color: #fcf8e3; color: #8a6d3b;">
  <strong>Primeira conclusão - Avaliações:</strong> <br>
  A maneira como o gráfico foi construído e como a análise foi feita não permitiu chegar no resultado esperado, uma vez que só indicou que eu tenho mais vendas com boa avaliação do que vendas com avaliações ruins. Então vou transformar a pergunta feita e considerar a média das avaliações dos produtos para tentar fazer uma análise similar
</div>

```python
#Apenas para seguir com a avaliação, aproveitei o dataframe para ver a distribuição das vendas e avaliações ao longo do tempo

df_resumo = df_vendas_versus_avaliação.groupby(['purchase_month', 'review_classification'])['order_id'].count().reset_index()
df_resumo.columns = ['purchase_month', 'review_classification', 'total_vendas']

# Definir a paleta de cores
palette = {
    'Boa': 'green',    # Cor verde para 'Boa'
    'Ruim': 'red',     # Cor vermelha para 'Ruim'
    'Neutro': 'orange' # Cor laranja para 'Neutra'
}

plt.figure(figsize=(15, 5))
sns.lineplot(data=df_resumo, x='purchase_month', y='total_vendas', hue='review_classification', palette=palette)
plt.xlabel('Mês de Compra')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Classificação de Avaliação ao Longo do Tempo')
plt.xticks(rotation=45)  
plt.grid(True)

plt.legend(title='Classificação de Avaliação')
plt.show()
```

<div style="border: 2px solid #f0ad4e; padding: 10px; border-radius: 5px; background-color: #fcf8e3; color: #8a6d3b;">
  <strong>Segunda conclusão - Avaliações:</strong> <br>
  A distribuição ao longo dos meses de vendas ficou interessante, mostra uma crescente em termos de avaliações positivas, o que é interessante para verificar se estão sendo implementadas ações de melhoria e atestar a efetividade delas. Da mesma forma a queda das avaliações negativas depois de julho de 2018 poderia ser melhor investigada, para verificar se foram ações de melhoria implementadas ou falha de registro de dados.<br>
  Interessante também analisar que as 3 curvas se comportam de maneira similar mais ou menos por volta de novembro de 2017.

```python
#criar um dataframe que unifique product_id pela sua média em review_score
df_media_notas = df_vendas_versus_avaliação.groupby("product_id")["review_score"].mean().reset_index()
df_media_notas.loc[df_media_notas['review_score'] <=2, 'review_classification'] = 'Ruim'
df_media_notas.loc[df_media_notas['review_score'] == 3, 'review_classification'] = 'Neutro'
df_media_notas.loc[df_media_notas['review_score'] >=4, 'review_classification'] = 'Boa'
df_media_notas.columns = ['product_id', 'media_nota', 'review_classification']

df_media_notas
```

```python
# Contar o número de vendas por produto e mês
df_vendas_mensais = df_vendas_versus_avaliação.groupby(['product_id', 'purchase_date'])['order_id'].count().reset_index()
df_vendas_mensais.columns = ['product_id', 'purchase_date', 'total_vendas']
df_vendas_mensais
```

```python
#juntando ambos os dataframes gerados
df_avaliacoes_completo = pd.merge(df_vendas_mensais, df_media_notas, on='product_id')
df_avaliacoes_completo
```

```python
df_avaliacoes_completo.total_vendas.sum()
```

```python
df_avaliacoes_completo.to_csv("datasets_criados/df_avaliacoes_completo.csv", index=False)
```

```python
#Travei na interpretação dos dados porque minhas tentativas de gráfico não davam certo e recorri ao Gemini para me ajudar a ordenar as ideias. Ele sugeriu gerar um novo dataframe, sendo esse abaixo
df_grouped = df_avaliacoes_completo.groupby('product_id').agg({'total_vendas': 'sum', 'media_nota': 'mean'}).reset_index()
```

```python
# Ajuste o tamanho da figura
plt.figure(figsize=(10, 6))

# Crie o gráfico de dispersão com uma paleta de cores mais intuitiva e uma grade para facilitar a leitura
sns.scatterplot(x='media_nota', y='total_vendas', data=df_grouped, hue="media_nota", palette='viridis', edgecolor='w')

# Adicione uma grade para melhorar a legibilidade dos pontos
plt.grid(True)

# Ajuste os rótulos dos eixos e o título
plt.xlabel('Média das Avaliações')
plt.ylabel('Total de Vendas')
plt.title('Relação entre Avaliação e Vendas')
plt.yticks(range(0, int(df_grouped['total_vendas'].max()) + 50, 50))
# Exiba o gráfico
plt.show()
```

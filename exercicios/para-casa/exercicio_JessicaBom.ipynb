Para isso foi necessÃ¡rio fazer uma junÃ§Ã£o entre os datasets: "olist_order_reviews_dataset" - que traz as avaliaÃ§Ãµes + "olist_orders_dataset" - que traz informaÃ§Ãµes das vendas + "olist_order_items" - que correlaciona o id do produto com o id do pedido.

```python
#Listando as colunas do df_pedidos para otimizar o drop
df_pedidos.columns
```

```python
#nosso df_pedidos jÃ¡ foi chamado acima e jÃ¡ tratamos a data da compra que Ã© a principal informaÃ§Ã£o que vamos usar, entÃ£o, em um primeiro momento nÃ£o vou precisar trabalhar nele, entÃ£o posso dropar as colunas que nÃ£o vou usar

columns_pedidos_drop = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date','order_estimated_delivery_date']
df_vendas_versus_avaliaÃ§Ã£o  = df_pedidos.drop(columns_pedidos_drop, axis=1)
df_vendas_versus_avaliaÃ§Ã£o
```

```python
df_avaliacoes_tratado = pd.read_csv("../../dataset_olist/dados brutos/olist_order_reviews_dataset.csv")
df_avaliacoes_tratado.columns
```

```python
#agora vamos comeÃ§ar a mergear com as informaÃ§Ãµes Ãºteis dos outros datasets, mas primeiro dropando deles o que nÃ£o Ã© Ãºltil para essa anÃ¡license

columns_avaliacoes_drop = ['review_comment_title', 'review_comment_message', 'review_answer_timestamp']
df_avaliacoes_tratado = df_avaliacoes_tratado.drop(columns_avaliacoes_drop, axis=1)
df_avaliacoes_tratado.dtypes
```

```python
#passando a data para datetime
df_avaliacoes_tratado['review_creation_date'] = pd.to_datetime(df_avaliacoes_tratado['review_creation_date'])
df_avaliacoes_tratado['review_date'] = df_avaliacoes_tratado['review_creation_date'].dt.date
df_avaliacoes_tratado['review_month'] = df_avaliacoes_tratado['review_creation_date'].dt.strftime('%Y-%m-%01')
df_avaliacoes_tratado['review_date'] = pd.to_datetime(df_avaliacoes_tratado['review_date'])
df_avaliacoes_tratado['review_month'] = pd.to_datetime(df_avaliacoes_tratado['review_month'])
df_avaliacoes_tratado.dtypes



```

```python
#agora vamos fazer o primeiro merge entre elas, pelo order_id

df_vendas_versus_avaliaÃ§Ã£o = pd.merge(df_vendas_versus_avaliaÃ§Ã£o, df_avaliacoes_tratado, on='order_id', how='inner')
df_vendas_versus_avaliaÃ§Ã£o
```

```python
#agora vamos fazer o tratamento da tabela que permite relacionar produto com pedido

df_produtos_tratado = pd.read_csv('../../dataset_olist/dados brutos/olist_order_items_dataset.csv')
df_produtos_tratado.columns
```

```python
columns_produtos_drop = ['order_item_id', 'seller_id',
       'shipping_limit_date', 'price', 'freight_value']

df_produtos_tratado = df_produtos_tratado.drop(columns_produtos_drop, axis=1)
df_produtos_tratado.dtypes
```

```python
#Vamos mergear agora esse df para unificar e comeÃ§ar as anÃ¡lises

df_vendas_versus_avaliaÃ§Ã£o = pd.merge(df_vendas_versus_avaliaÃ§Ã£o, df_produtos_tratado, on='order_id', how='inner')
df_vendas_versus_avaliaÃ§Ã£o
```

```python
#Vou criar uma coluna de classificaÃ§Ã£o (review_classification) com base na nota da avaliaÃ§Ã£o (review_score)

df_vendas_versus_avaliaÃ§Ã£o.loc[df_vendas_versus_avaliaÃ§Ã£o['review_score'] <=2, 'review_classification'] = 'Ruim'
df_vendas_versus_avaliaÃ§Ã£o.loc[df_vendas_versus_avaliaÃ§Ã£o['review_score'] == 3, 'review_classification'] = 'Neutro'
df_vendas_versus_avaliaÃ§Ã£o.loc[df_vendas_versus_avaliaÃ§Ã£o['review_score'] >=4, 'review_classification'] = 'Boa'
df_vendas_versus_avaliaÃ§Ã£o
```

```python
#Aqui usei um pouco o chatGPT para ajudar a entender a lÃ³gica a ser aplicada para construir esse novo DF para fazer a avaliaÃ§Ã£o
# 1. Agrupar os dados por 'review_classification' e contar as vendas (order_id)
df_resumo = df_vendas_versus_avaliaÃ§Ã£o.groupby('review_classification')['order_id'].count().reset_index()

# 2. Renomear a coluna para 'total_vendas' para mais clareza
df_resumo.columns = ['review_classification', 'total_vendas']

# 3. Plotar o grÃ¡fico de barras
plt.figure(figsize=(8, 6))
plt.bar(df_resumo['review_classification'], df_resumo['total_vendas'], color=['green', 'orange', 'red'])
plt.xlabel('ClassificaÃ§Ã£o de AvaliaÃ§Ã£o')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por ClassificaÃ§Ã£o de AvaliaÃ§Ã£o')
plt.xticks(rotation=0)
plt.show()
```

  A maneira como o grÃ¡fico foi construÃ­do e como a anÃ¡lise foi feita nÃ£o permitiu chegar no resultado esperado, uma vez que sÃ³ indicou que eu tenho mais vendas com boa avaliaÃ§Ã£o do que vendas com avaliaÃ§Ãµes ruins. EntÃ£o vou transformar a pergunta feita e considerar a mÃ©dia das avaliaÃ§Ãµes dos produtos para tentar fazer uma anÃ¡lise similar

```python
#Apenas para seguir com a avaliaÃ§Ã£o, aproveitei o dataframe para ver a distribuiÃ§Ã£o das vendas e avaliaÃ§Ãµes ao longo do tempo

df_resumo = df_vendas_versus_avaliaÃ§Ã£o.groupby(['purchase_month', 'review_classification'])['order_id'].count().reset_index()
df_resumo.columns = ['purchase_month', 'review_classification', 'total_vendas']

# Definir a paleta de cores
palette = {
    'Boa': 'green',    # Cor verde para 'Boa'
    'Ruim': 'red',     # Cor vermelha para 'Ruim'
    'Neutro': 'orange' # Cor laranja para 'Neutra'
}

plt.figure(figsize=(15, 5))
sns.lineplot(data=df_resumo, x='purchase_month', y='total_vendas', hue='review_classification', palette=palette)
plt.xlabel('MÃªs de Compra')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por ClassificaÃ§Ã£o de AvaliaÃ§Ã£o ao Longo do Tempo')
plt.xticks(rotation=45)  
plt.grid(True)

plt.legend(title='ClassificaÃ§Ã£o de AvaliaÃ§Ã£o')
plt.show()
```

```python
#criar um dataframe que unifique product_id pela sua mÃ©dia em review_score
df_media_notas = df_vendas_versus_avaliaÃ§Ã£o.groupby("product_id")["review_score"].mean().reset_index()
df_media_notas.loc[df_media_notas['review_score'] <=2, 'review_classification'] = 'Ruim'
df_media_notas.loc[df_media_notas['review_score'] == 3, 'review_classification'] = 'Neutro'
df_media_notas.loc[df_media_notas['review_score'] >=4, 'review_classification'] = 'Boa'
df_media_notas.columns = ['product_id', 'media_nota', 'review_classification']

df_media_notas
```

```python
# Contar o nÃºmero de vendas por produto e mÃªs
df_vendas_mensais = df_vendas_versus_avaliaÃ§Ã£o.groupby(['product_id', 'purchase_date'])['order_id'].count().reset_index()
df_vendas_mensais.columns = ['product_id', 'purchase_date', 'total_vendas']
df_vendas_mensais
```

```python
#juntando ambos os dataframes gerados
df_avaliacoes_completo = pd.merge(df_vendas_mensais, df_media_notas, on='product_id')
df_avaliacoes_completo
```

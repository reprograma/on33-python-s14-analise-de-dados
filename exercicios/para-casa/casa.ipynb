```python
# Unir df_order com df_pagamento
df_order_payment = pd.merge(df_order, df_pagamento, on='order_id', how='inner')

# Unir o resultado com df_itens
df_full = pd.merge(df_order_payment, df_itens, on='order_id', how='inner')

# Remover as duas Ãºltimas colunas pelo nome que constam valores nulos
df_full = df_full.drop(columns=['Unnamed: 7', 'Unnamed: 8'])

# Visualizar a base final
df_full.head()
```

```python
# Verificar valores ausentes
df_full.isnull().sum()

# Remover duplicatas
df_full = df_full.drop_duplicates()
```

```python
print(df_full.isnull().sum())

df_full = df_full.drop_duplicates()

print("\nDescriÃ§Ã£o estatÃ­stica dos dados apÃ³s remoÃ§Ã£o de duplicatas:")
print(df_full.describe())
```

```python
#Qual o top  dos vendedores com maior nÃºmero de vendas?

# Agrupar por 'seller_id' e somar o 'payment_value'
vendedores_vendas = df_full.groupby('seller_id')['payment_value'].sum().reset_index()

# Renomear a coluna para mais clareza
vendedores_vendas = vendedores_vendas.rename(columns={'payment_value': 'total_sales_value'})

# Ordenar os vendedores com base no valor total das vendas em ordem decrescente
vendedores_vendas = vendedores_vendas.sort_values(by='total_sales_value', ascending=False)

# Mostrar os vendedores com mais vendas
print(vendedores_vendas.head())
```

```python
#Qual os  produtos com maior Ã­ndice de vendas?

# Agrupar por 'product_id' e somar o nÃºmero de itens vendidos
produtos_vendas = df_full.groupby('product_id')['order_item_id'].count().reset_index()

# Renomear a coluna para mais clareza
produtos_vendas = produtos_vendas.rename(columns={'order_item_id': 'total_items_sold'})

# Ordenar os produtos com base na quantidade total de itens vendidos em ordem decrescente
produtos_vendas = produtos_vendas.sort_values(by='total_items_sold', ascending=False)

# Mostrar os produtos mais vendidos
print(produtos_vendas.head())

```

```python
#Qual Ã© a distribuiÃ§Ã£o de vendas por tipo de pagamento?

# Agrupar por 'payment_type' e somar o 'payment_value'
vendas_por_pagamento = df_full.groupby('payment_type')['payment_value'].sum().reset_index()

# Renomear a coluna para mais clareza
vendas_por_pagamento = vendas_por_pagamento.rename(columns={'payment_value': 'total_sales_value'})

# Ordenar por valor total de vendas em ordem decrescente
vendas_por_pagamento = vendas_por_pagamento.sort_values(by='total_sales_value', ascending=False)

# Mostrar a distribuiÃ§Ã£o das vendas por tipo de pagamento
print(vendas_por_pagamento)
```

```python
 #Qual Ã© a relaÃ§Ã£o entre a pontuaÃ§Ã£o das avaliaÃ§Ãµes e o valor total das vendas?
 
 # Agrupar por 'review_score' e somar o 'payment_value'
avaliacoes_vendas = df_full.groupby('review_score')['payment_value'].sum().reset_index()

# Renomear a coluna para mais clareza
avaliacoes_vendas = avaliacoes_vendas.rename(columns={'payment_value': 'total_sales_value'})

# Ordenar por 'review_score' para visualizar a relaÃ§Ã£o
avaliacoes_vendas = avaliacoes_vendas.sort_values(by='review_score')

# Mostrar a relaÃ§Ã£o entre pontuaÃ§Ã£o das avaliaÃ§Ãµes e valor total das vendas
print(avaliacoes_vendas)
```

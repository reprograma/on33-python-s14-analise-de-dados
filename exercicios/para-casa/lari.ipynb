```python
# Tratando nulos

products = products.dropna()
print(products.isnull().sum())
order_items['freight_value'] = order_items['freight_value'].fillna(0)
```

```python
# Tratando as datas
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders.info()
```

```python
# Criando nova coluna: calculando o valor total de cada pedido
order_items['total_value'] = order_items['price'] * order_items['order_item_id']
order_items.head()
```

```python
# Pergunta 1: Qual é o valor médio gasto por cliente?

# Unindo dados de pedidos e itens de pedido para calcular o valor total gasto por cliente
orders_items_merged = pd.merge(orders, order_items, on='order_id')

# Agrupar por cliente e calcular o valor total gasto por cliente
valor_gasto_por_cliente = orders_items_merged.groupby('customer_id')['price'].sum()

# Verificar o valor médio gasto por cliente
valor_medio_gasto = valor_gasto_por_cliente.mean()
print(f'Valor médio gasto por cliente: R${valor_medio_gasto:.2f}')
```

```python
# Pergunta 2: Qual é o estado com o maior número de pedidos?

# Unir dados de clientes e pedidos
clientes_pedidos = pd.merge(customers, orders, on='customer_id')

# Agrupar por estado e contar o número de pedidos
pedidos_por_estado = clientes_pedidos.groupby('customer_state')['order_id'].count().sort_values(ascending=False)

# Exibir o estado com o maior número de pedidos
estado_com_mais_pedidos = pedidos_por_estado.idxmax()
print(f'O estado com o maior número de pedidos é: {estado_com_mais_pedidos}')
```

```python
# Pergunta 3: Qual é o valor médio dos produtos por categoria?

# Unir os dados de produtos e itens de pedido
produtos_itens = pd.merge(products, order_items, on='product_id')

# Calcular o preço médio por categoria de produto
preco_medio_por_categoria = produtos_itens.groupby('product_category_name')['price'].mean().sort_values(ascending=False)

# Exibir o preço médio por categoria
preco_medio_por_categoria.head()
```

```python
# Pergunta 4: Qual é a relação entre o número de itens e o valor total de um pedido?

# Agrupar por pedido e calcular o número de itens e o valor total de cada pedido
itens_valor_por_pedido = orders_items_merged.groupby('order_id').agg({'order_item_id': 'count', 'price': 'sum'})

# Exibir a relação entre o número de itens e o valor total de um pedido
itens_valor_por_pedido.plot(kind='scatter', x='order_item_id', y='price', figsize=(10,6))
plt.xlabel('Número de Itens por Pedido')
plt.ylabel('Valor Total do Pedido')
plt.title('Relação entre Número de Itens e Valor Total do Pedido')
plt.show()
```

```python
# Gráfico de barras do número de pedidos por estado
pedidos_por_estado.plot(kind='bar', figsize=(12,6))
plt.xlabel('Estado')
plt.ylabel('Número de Pedidos')
plt.title('Número de Pedidos por Estado')
plt.show()
```

```python
# Distribuição dos pedidos ao longo do tempo

# Converter a coluna de data para o formato datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Criar uma nova coluna com o mês e ano do pedido
orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')

# Agrupar por ano e mês e contar o número de pedidos
pedidos_por_tempo = orders.groupby('year_month')['order_id'].count()

# Criar um gráfico de linha para mostrar a evolução dos pedidos ao longo do tempo
pedidos_por_tempo.plot(kind='line', figsize=(10,6), marker='o')
plt.xlabel('Ano e Mês')
plt.ylabel('Número de Pedidos')
plt.title('Evolução do Número de Pedidos ao Longo do Tempo')
plt.grid(True)
plt.show()
```

```python
# Qual é o valor médio dos pedidos por estado?

# Unir dados de clientes, pedidos e itens de pedidos
clientes_pedidos_items = pd.merge(pd.merge(customers, orders, on='customer_id'), order_items, on='order_id')

# Agrupar por estado e calcular o valor médio dos pedidos
valor_medio_por_estado = clientes_pedidos_items.groupby('customer_state')['price'].mean().sort_values(ascending=False)

# Exibir o valor médio por estado em um gráfico de barras horizontais
valor_medio_por_estado.plot(kind='barh', figsize=(12,8))
plt.xlabel('Valor Médio (R$)')
plt.ylabel('Estado')
plt.title('Valor Médio dos Pedidos por Estado')
plt.show()
```

```python
# Qual categoria de produto gera o maior valor total de vendas?

# Unir os dados de produtos e itens de pedidos
produtos_vendas = pd.merge(products, order_items, on='product_id')

# Agrupar por categoria de produto e calcular o valor total de vendas
valor_total_por_categoria = produtos_vendas.groupby('product_category_name')['price'].sum().sort_values(ascending=False)

# Exibir o valor total por categoria em um gráfico de barras
valor_total_por_categoria.plot(kind='bar', figsize=(12,8))
plt.xlabel('Categoria')
plt.ylabel('Valor Total de Vendas (R$)')
plt.title('Valor Total de Vendas por Categoria de Produto')
plt.xticks(rotation=90)
plt.show()
```

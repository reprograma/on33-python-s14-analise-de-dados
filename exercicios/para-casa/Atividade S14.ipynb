```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_order_items = pd.read_csv('./olist_order_items_dataset.csv')  # Itens dos pedidos
df_order_items.head(10)
```

```python
df_order_items.isnull().sum()
print(df_order_items)
```

```python
#Removendo duplicadas
df_order_items.drop_duplicates(inplace=True)
```

```python
print(df_order_items.dtypes)
```

```python
df_review = pd.read_csv('./olist_order_reviews_dataset.csv')  # Produtos
df_review.head(10)
```

```python
df_review.isnull().sum()
print(df_review)
```

```python
df_review.drop_duplicates(inplace=True)
```

```python
print(df_review.dtypes)
```

```python
df_products = pd.read_csv('./olist_products_dataset.csv')  # Avaliações dos pedidos
df_products.head(10)
```

```python
df_products.isnull().sum()
print(df_products)
```

```python
df_products.drop_duplicates(inplace=True)
```

```python
print(df_products.dtypes)
```

```python
#juntando as bases
df_base_completa = pd.merge(df_review, df_order_items, on="order_id", how="inner")
df_base_completa = pd.merge(df_products, df_base_completa, on="product_id", how="inner")
df_base_completa.head()
```

```python
print(df_base_completa.dtypes)
```

```python
#Salvando nova base

df_base_completa.to_csv('Olist_base_completa.csv', index=False)
```

Pergunta 1: Quais são as categorias de produtos mais vendidas?

```python
categorias_mais_vendidas = df_base_completa.groupby('product_category_name').size().reset_index(name='num_pedidos').sort_values(by='num_pedidos', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=categorias_mais_vendidas.head(10), x='product_category_name', y='num_pedidos', palette='Blues_d')
plt.title('Top 10 Categorias de Produtos Mais Vendidas')
plt.xlabel('Categoria de Produto')
plt.ylabel('Número de Pedidos')
plt.xticks(rotation=45)
plt.show()
```

Análise: O gráfico exibe as 10 categorias de produtos mais vendidas, ordenadas pelo número de pedidos. Observa-se que categorias como Cama, Mesa e Banho, Beleza e Saude, e Esporte e Lazer se destacam em termos de volume de vendas. Essa tendência pode indicar uma forte demanda por produtos que atendem às necessidades diárias dos consumidores, mas também de cuidados. As categorias que aparecem no topo sugerem que os consumidores priorizam itens que oferecem funcionalidade e conforto.

Pergunta 2: Qual é o valor médio dos produtos por categoria?

```python
#top 10 categorias
top_categorias = categorias_mais_vendidas.head(10)['product_category_name']
```

```python
base_top_categorias = df_base_completa[df_base_completa['product_category_name'].isin(top_categorias)]
valor_medio_categoria = base_top_categorias.groupby('product_category_name')['price'].mean().reset_index().sort_values(by='price', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=valor_medio_categoria, x='product_category_name', y='price', palette='Greens_d')
plt.title('Distribuição dos Valores dos Produtos por Categoria (Top 10)')
plt.xlabel('Categoria de Produto')
plt.ylabel('Valor do Produto (R$)')
plt.xticks(rotation=45)
plt.show()
```

Análise: O gráfico de boxplot ilustra a distribuição dos preços dos produtos nas 10 categorias mais vendidas. Notam-se diferenças significativas nos preços médios entre as categorias. Por exemplo, Relógios e Automóveis tendem a apresentar valores mais altos, enquanto categorias como Telefonia possuem preços médios mais baixos. 

Pergunta 3: Qual é a distribuição das avaliações dos clientes por categoria de produto?

```python
base_top_categorias = df_base_completa[df_base_completa['product_category_name'].isin(top_categorias)]
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=base_top_categorias, x='product_category_name', y='review_score', palette='coolwarm', inner='quartile')
plt.title('Distribuição das Avaliações por Categoria de Produto (Top 10)')
plt.xlabel('Categoria de Produto')
plt.ylabel('Nota de Avaliação')
plt.xticks(rotation=45)
plt.show()
```

Análise: O gráfico de violino mostra a distribuição das notas de avaliação por categoria de produto. A maioria das categorias apresenta uma concentração de avaliações ao redor de notas Altas, indicando que, em geral, os consumidores têm uma percepção positiva em relação aos produtos adquiridos. Essa informação é útil para identificar áreas que podem necessitar de melhorias na qualidade dos produtos ou no atendimento ao cliente.

Pergunta 4: Qual é a relação entre a nota de avaliação e o valor do produto?

```python
nota_valor_produto = df_base_completa.groupby('review_score').agg({
    'price': 'mean'
}).reset_index()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=nota_valor_produto, x='review_score', y='price', marker='o', palette='viridis')
plt.title('Relação entre Nota de Avaliação e Valor Médio dos Produtos')
plt.xlabel('Nota de Avaliação')
plt.ylabel('Valor Médio (R$)')
plt.show()
```

Análise: O gráfico de linha representa a relação entre a nota de avaliação e o valor médio dos produtos. Observa-se uma tendência de que produtos com avaliações mais altas tendem a ter preços mais intermediário. Isso sugere que os consumidores estão dispostos a pagar mais por produtos que oferecem qualidade superior. Contudo, é importante notar que essa relação não é linear; existem produtos com baixa avaliação que tem os maiores preços. Essa análise destaca a importância de manter a qualidade do produto e o atendimento ao cliente, pois eles impactam diretamente na percepção de valor e satisfação do consumidor.

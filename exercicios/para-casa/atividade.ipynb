```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
sns.histplot(df_olist["payment_value_x"], binwidth=50);
```

```python
df_sazonalidade = df_olist.groupby('payment_value_x').agg({'order_id': 'nunique'})
df_sazonalidade.sort_values(by='order_id', ascending=False)
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df_sazonalidade.reset_index().sort_values("payment_value_x"), x="payment_value_x", y="order_id")
plt.xticks(rotation=45)
```

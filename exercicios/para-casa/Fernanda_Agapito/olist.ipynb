## Atividade para casa 

```python
df_join.columns
```

```python
df_avaliacoes = pd.read_csv('./olist/olist_order_reviews_dataset.csv')
df_avaliacoes.head()
```

```python
df_join.columns
```

```python
df_join = pd.merge(df_join, df_avaliacoes, on="order_id", how="inner")
df_join.head()
```

```python
df_vendedores = pd.read_csv('./olist/olist_sellers_dataset.csv')
df_vendedores.head()
```

```python
df_join = pd.merge(df_join, df_vendedores, on="seller_id", how="inner")
df_join.head()
```

```python
# Agrupar as vendas por produto
produtos_vendas = df_join.groupby('product_id').agg({'order_item_id': 'count', 'price': 'sum'}).reset_index()
produtos_vendas.columns = ['product_id', 'total_vendas', 'total_revenue']

# Agrupar as avaliações por produto (extraindo do df_join)
produtos_reviews = df_join.groupby('product_id').agg({'review_score': 'mean'}).reset_index()
produtos_reviews.columns = ['product_id', 'avg_review']

# Mesclar vendas com avaliações
produtos_info = produtos_vendas.merge(produtos_reviews, on='product_id')

# Mesclar com a localização dos vendedores (estado e cidade)
produtos_info = produtos_info.merge(df_join[['product_id', 'seller_state', 'seller_city']].drop_duplicates(), on='product_id')
```

```python
# Ordenar produtos por total de vendas e avaliação média
produtos_mais_vendidos = produtos_info.sort_values(by=['total_vendas', 'avg_review'], ascending=[False, False])
produtos_mais_vendidos.head(10)  # Mostrar os 10 produtos mais vendidos com melhores avaliações
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 produtos
top_produtos = produtos_mais_vendidos.head(10)

# Gráfico de barras das vendas e avaliação
plt.figure(figsize=(15,6))
sns.barplot(x='product_id', y='total_vendas', data=top_produtos, color='blue', label='Total Vendas')
sns.barplot(x='product_id', y='avg_review', data=top_produtos, color='orange', label='Média Avaliações')
plt.xticks(rotation=45)
plt.legend()
plt.title('Top 10 Produtos com Mais Vendas e Melhores Avaliações')
plt.show()
```

```python
# Agrupar por estado
vendas_por_estado = produtos_info.groupby('seller_state')['total_vendas'].sum().reset_index()

# Plotar o gráfico de barras

plt.figure(figsize=(8,4))
sns.lineplot(data=vendas_por_estado, x='seller_state', y='total_vendas')
plt.title('Total de Vendas por Estado')
plt.xlabel('Estado')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```

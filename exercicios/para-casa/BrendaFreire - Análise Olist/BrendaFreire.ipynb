Plotando os gráficos para as análises e testes de hipóteses

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Qual a relação entre o valor do frete e o valor do pagamento?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='freight_value', y='payment_value', data=df_lista_ETL, alpha=0.5)
plt.title('Relação entre Valor do Frete e Valor do Pagamento')
plt.xlabel('Valor do Frete (freight_value)')
plt.ylabel('Valor do Pagamento (payment_value)')
plt.grid(True)
plt.show()
```

Nesta primeira pergunta, pode-se perceber uma concentração de compras efetuadas quando o valor do frete é menor, fazendo com que o cliente finalize a sua compra

2. Produtos mais caros recebem melhores ou piores avaliações?

```python
# Categorizar produtos em baratos e caros com base na mediana do preço
df_lista_ETL['price_category'] = ['Alto' if price > df_lista_ETL['price'].mean() else 'Baixo' for price in df_lista_ETL['price']]

# Gráfico de barras comparando categorias de preço e média de review_score
plt.figure(figsize=(8, 6))
sns.barplot(x='price_category', y='review_score', data=df_lista_ETL, estimator=lambda x: x.mean())
plt.title('Média das Avaliações por Categoria de Preço')
plt.xlabel('Categoria de Preço')
plt.ylabel('Média da Avaliação (review_score)')
plt.show()
```

Aparentemente, o valor do produto não interfere na avaliação do cliente, tendo em vista que produtos com preço acima da média e produtos abaixo da média ainda assim recebem avaliações

3. Qual o método de pagamento mais utilizado?

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='payment_type', data=df_lista_ETL, palette='viridis')
plt.title('Método de Pagamento Mais Utilizado')
plt.xlabel('Método de Pagamento')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

Com base na informação de que o cartão de crédito é o método mais utilizado pelos clientes para efetuar as compras, pergunta-se:

4. O método de pagamento influencia a pontuação das avaliações?

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='payment_type', y='review_score', data=df_lista_ETL, estimator=lambda x: x.mean(), palette='viridis')
plt.title('Média das Avaliações por Método de Pagamento')
plt.xlabel('Método de Pagamento')
plt.ylabel('Média da Avaliação (review_score)')
plt.xticks(rotation=45)
plt.show()
```

Como demonstrado pelo gráfico criado, independentemente do método de pagamento, as avaliações dos produtos mantém uma média boa e parecida quando comparados os payment types

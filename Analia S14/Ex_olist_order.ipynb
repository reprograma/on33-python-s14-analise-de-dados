```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('olist_order_payments_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

1-  Qual a distribuição para os tipos de pagamentos?

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='payment_type', palette='colorblind')
plt.title('Distribuição dos Tipos de Pagamento', fontsize=16)
plt.xlabel('Tipo de Pagamento', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

2- Qual média do valor de pagamento por tipo de pagamento?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='payment_type', y='payment_value', palette='colorblind', ci=None)
plt.title('Média do Valor de Pagamento por Tipo de Pagamento', fontsize=16)
plt.xlabel('Tipo de Pagamento', fontsize=12)
plt.ylabel('Valor Médio do Pagamento (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

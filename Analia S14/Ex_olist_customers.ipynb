```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1- Qual é a distribuição geográfica dos clientes?

```python
# Gráfico de barras que mostre o número de clientes por estado ou cidade

plt.figure(figsize=(12, 6))
df['customer_state'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição de Clientes por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=45)
plt.show()

```

2- Quantos clientes únicos existem em cada estado?

```python
# Gráfico de barras que mostra o número de clientes únicos por estado

clientes_unicos_por_estado = df.groupby('customer_state')['customer_unique_id'].nunique()

plt.figure(figsize=(12, 6))
clientes_unicos_por_estado.plot(kind='bar', color='purple')
plt.title('Número de Clientes Únicos por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Clientes Únicos')
plt.xticks(rotation=45)
plt.show()


```
